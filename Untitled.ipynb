```python
def find_zero_indexes(arr):
    zero_indexes = []
    for i, num in enumerate(arr):
        if num == 21.583198773968626:
            zero_indexes.append(i)
    return zero_indexes

# Example usage:
zero_indices = find_zero_indexes(s.ll)
if zero_indices:
    print("The array returns to 0 at indexes:", zero_indices)
else:
    print("0 is not found in the array.")
```
